```python
import cv2
import os
import pydicom

outdir = './PNG-Images'
os.makedirs(outdir, exist_ok=True)

# Step 1
#Specify the dataPath 
mainFolderPath = r'D:\Teknofest\Inference\Inference\competition_dataset\competition_dataset'

# Read Folders
folderPaths = [os.path.join(mainFolderPath, f) for f in os.listdir(mainFolderPath)]
pIDS = [f for f in os.listdir(mainFolderPath)]
print("Length of the folders: ", len(folderPaths))

for i in range(len(pIDS)):
    out = os.path.join(outdir, pIDS[i])
    os.makedirs(out, exist_ok=True)
    files = os.listdir(os.path.join(mainFolderPath, pIDS[i]))
    for f in files:
        filepath = os.path.join(mainFolderPath, pIDS[i], f)
        print(filepath)
        ds = pydicom.dcmread(filepath)
        
        # Extract the image array from the DICOM file
        img = ds.pixel_array
        
        # Normalize image if necessary (convert to 8-bit grayscale)
        img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
        
        # Show the image (optional)
        # cv2.imshow('image', img)
        # cv2.waitKey(0)
        # cv2.destroyAllWindows()

        # Save the image as PNG
        cv2.imwrite(os.path.join(out, f.replace('.dcm', '.png')), img)
```
